```python
import requests
from bs4 import BeautifulSoup

url="https://comic.naver.com/webtoon/list.nhn?titleId=20853&weekday=tue"
shtml = requests.get(url).text
soup = BeautifulSoup(shtml, "html.parser")

outDir = os.getcwd() + "/images"  

selector = "tr>td>a>img"
for tag in soup.select(selector):
    filename = tag["src"]
    filename = outDir + "/" +  os.path.split(filename)[1]
    
    with open(filename, "wb") as f:
        image = requests.get(tag["src"]).content
        f.write( image )
        print("이미지저장....{}..{}".format(filename,len(image)))
    
```

```python
from IPython.display import Image, display
import glob

imglist = glob.glob(outDir + "/*.jpg")

for img in imglist:
    display(Image(filename=img, width=100, height=100))
```

```python
from selenium import webdriver
import time

browser = webdriver.Chrome("./driver/chromedriver.exe") #실행 
browser.get("https://www.w3schools.com/colors/colors_rgb.asp")  #특정사이트 

time.sleep(2)

browser.find_element_by_css_selector("#rgb01").clear()

browser.find_element_by_css_selector("#r01").clear()
browser.find_element_by_css_selector("#g01").clear()
browser.find_element_by_css_selector("#b01").clear()

browser.find_element_by_css_selector("#r01").send_keys("0")
browser.find_element_by_css_selector("#g01").send_keys("255")
browser.find_element_by_css_selector("#b01").send_keys("255")

#browser.find_element_by_css_selector("#rgb01")
html = browser.page_source
soup = BeautifulSoup(html, "html.parser")
hexa = soup.find("div", {"id":"hex01"})
print(hexa)

time.sleep(5)
browser.quit()
```
